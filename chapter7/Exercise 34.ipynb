  In  this  exercise,  I  want  you  to  write  a get_sv  function  that  returns  a  set  of  all“supervocalic”  words  in  the  dict.  If  you’ve  never  heard  the  term supervocalic  before,you’re not alone: I only learned about such words several years ago. Simply put, suchwords  contain  all  five  vowels  in  English  (a,  e,  i,  o,  and  u),  each  of  them  appearingonce and in alphabetical order.

```python
def get_sv(filename):
    vowels=set('aeiou')
    return {word.strip()
            for word in open(filename)
            if vowels<set(word.lower())
    }
get_sv("words.txt")
```

In the /etc/passwd  file you  used earlier, what  different  shells  (i.e.,  commandinterpreters, named in the final field on each line) are assigned to users? Use aset comprehension to gather them.

```python
def gather_shells(filename):
    return {
        shell.strip().split()[-1]
        for shell in open(filename)
        if not shell.startswith('#')
    }
gather_shells('passwd.txt')
```
